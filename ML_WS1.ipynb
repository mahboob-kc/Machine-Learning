### NAIVE_

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
mydata = pd.read_csv("/Iris_data.csv")
mydata
```

```python
y=mydata.Species
y
```

```python
x=mydata.iloc[:,1:5]
x
```

```python
mymodel = GaussianNB()
mymodel.fit(x,y)
```

```python
accuracy = mymodel.score(x,y)
round(accuracy*100,2)
```

```python
mytest = pd.read_csv("/Iris_test.csv")
mytest
```

```python
yy=mytest.Species
yy
```

```python
xx=mytest.iloc[:,1:5]
xx
```

```python
accuracy1 = mymodel.score(xx,yy)
round(accuracy1*100,2)
```

```python
predtext = mymodel.predict(xx)
predtext
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(yy,predtext)
mat
```

```python
from sklearn.metrics import classification_report
report =  classification_report (yy,predtext)
print(report)
```

### linear regression

```python
import matplotlib.pyplot as myplot
import seaborn as mysb
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("/Mult_Reg_Yield.csv")
data
```

```python
data.describe()
```

```python
ey=data.Yield
ey
```

```python
ex=data.iloc[:,0:2]
ex
```

```python
mysb.pairplot(data)
myplot.show
```

```python
mymodel1=LinearRegression()
mymodel1.fit(ex,ey)
```

```python
mymodel1.intercept_
```

```python
mymodel1.coef_
```

```python
pred = mymodel1.predict(ex)
pred
```

```python
msd=mean_squared_error(ey,pred)
msd
```

```python
pred = pd.DataFrame(pred,columns=['predicted'])
pred
```
